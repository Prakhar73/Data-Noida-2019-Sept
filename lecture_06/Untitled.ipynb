```python
theta = np.linspace(0, 10 * np.pi, 1000)
```

```python
sin = np.sin(theta)
cos = np.cos(theta)
```

```python
plt.plot(theta, sin)
plt.plot(theta, cos)
```

```python
r = np.linspace(10, 20, 1000)
x_center, y_center = 100, 100
```

```python
x = r * cos + x_center
y = r * sin + y_center
```

```python
plt.plot(x, y)
```

```python
img = np.zeros([100, 100])
```

```python
center = (50, 50)
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        
        distance = np.sqrt((row-center[0])**2 + (col-center[1])**2)
        
        img[row, col] = distance % 20
```

```python
plt.imshow(img, cmap="gray")
```
