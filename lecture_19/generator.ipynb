```python
from keras.preprocessing.image import ImageDataGenerator, load_img
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = load_img("../datasets/panda.jpg")
```

```python
np_img = np.array(img)
```

```python
img_gen = ImageDataGenerator(rotation_range=30, shear_range=.5, horizontal_flip=True, zoom_range=.5)
```

```python
gen = img_gen.flow(np.array([np_img]))
```

```python
batch = gen.next()

for item in batch:
    plt.figure()
    plt.imshow(item.astype(int))
```
