```python
from keras.preprocessing.image import load_img
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = load_img("../datasets/panda.jpg", target_size=(280, 280), grayscale=True)
np_img = np.array(img)
```

```python
plt.imshow(np_img, cmap="gray")
```

```python
target = np.zeros((np_img.shape[0]-2, np_img.shape[1] - 2))
```

```python
fil = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
```

```python
for i in range(np_img.shape[0]-2):
    for j in range(np_img.shape[1]-2):
        
        cut = np_img[i:i+3, j:j+3]
        val = np.mean(cut*fil)
        target[i, j] = np.abs(val)
```

```python
plt.imshow(target.astype(int), cmap="gray")
```
